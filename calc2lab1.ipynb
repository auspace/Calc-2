```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): # define the function
    return 1 / np.sqrt(7 * x + 1)

x = np.linspace(0, 1, 100) # generate 100 points from 0 to 1
y = f(x) # compute f(x) for each point

plt.plot(x, y) # plot the function
plt.xlabel('x') # label x-axis
plt.ylabel('f(x)') # label y-axis
plt.grid(True) # add grid
plt.show() # display the plot

area = 2/7 * (np.sqrt(8) - 1)
print(f"area = {area}")
```

## Looking at the grid the only point of intersection is (0,1)

```python
import sympy as sp

x = sp.Symbol('x') # def variable

y1 = sp.exp(x) # def functions
y2 = sp.sqrt(7*x) + 1

area = sp.integrate(y2 - y1, (x, 0, 1)) # Compute area between curves from 0 to 1

print("area =", area)
```

```python
import sympy as sp

x = sp.Symbol('x') # define x

y1 = sp.exp(x) # def funcs
y2 = sp.sqrt(7*x) + 1

volume = 2*sp.pi * sp.integrate(x * (y2 - y1), (x, 0, 1)) # compute volume using shell method

print("volume =", volume)
```

## Derivitive of right side

```python
import sympy as sp

x, n = sp.symbols('x n') # def var
rs = sp.integrate((sp.ln(x))**(n-1), x) # right side of formula
derivative = sp.diff(rs, x) # derivitive
dsimplified = sp.simplify(derivative) # derivitive simplified

print("Derivative of right side", dsimplified)
```

## Inetegral of left side when n = 7

```python
import sympy as sp

x = sp.symbols('x') # def var
ls = sp.integrate(x*(sp.ln(x))**7 - 7)

print(f"Integral of left side when n = 7 equals {ls}") # integral of left side
```
