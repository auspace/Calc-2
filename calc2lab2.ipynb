```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, t = sp.symbols('x t') # Define symbols

h = (4*x**3 - 27*x**2 + 5*x - 32)/(30*x**5 - 13*x**4 + 50*x**3 - 286*x**2 - 299*x - 70) # Define function

pf = sp.apart(h, x) # partial fraction decomposition
print("Partial Fractions:", pf) 

int_pf = sp.integrate(pf, x) # integrate using partial fractions
print("\n(Partial Fractions) dx =", int_pf)

int_orig = sp.integrate(h, x) # integrate original function
print("\n(Original) dx =", int_orig)

print("\nDifference:", sp.simplify(int_pf - int_orig)) # check if both integrals are the same
```

## 2c: since g(x) is always smaller than 1/(x^2) and the integral of 1/(x^2) converges then g(x) must also converge.

```python
g = sp.cos(x)**2 / x**2 # Define g(x)
for val in [2,5,10,100,1000,10000]: # go through vals
    res = sp.integrate(g, (x, 1, val)) # Compute the integral from 1 to val
    print(f"\n integral g(x) dx from 1 to {val} =", res.evalf())

f = 1/x**2 # Define f(x)
xx = np.linspace(1,10,400) # x values for plotting
plt.plot(xx, np.cos(xx)**2/xx**2, label='g(x)') # plot g(x)
plt.plot(xx, 1/xx**2, label='1/x²', linestyle='--') # plot f(x)
plt.show() 
```

```python
a = np.e # Eulers num
y = sp.log(x, a) # log_a(x)

xx = np.linspace(1,2,100) # x values
yy = np.log(xx)/np.log(a) # y values
plt.plot(xx, yy); plt.show() 

dy = sp.diff(y, x) # dy/dx
L = sp.integrate(sp.sqrt(1 + dy**2), (x,1,2)) # arc length
print("\nArc length =", L.evalf())
```
