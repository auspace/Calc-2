## Problem 2

```python
# (a)
import matplotlib.pyplot as plt
import math

a = 16

xvals = list(range(21))
gvals = [ (x**a) / math.factorial(x) if x > 0 else 0 for x in xvals ]

plt.stem(xvals, gvals)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title(f'g(x) = x^{a} / x!')
plt.show()
```

### (b) Limit as x -> infinity:
### g(x) = x^a / x! 
### As x increases, x! grows faster than x^a
### Therefore, limit g(x) -> 0 as x -> infinity

```python
# (c)
a = 16
gx = 1 / a**3

n = 1
while True:
    g = n**a / math.factorial(n)
    if g < gx:
        N = n
        break
    n += 1

print("Smallest N =", N)
```

## Problem 3

```python
def calca(n):  
    return (math.log(n))**2 / n**2

# (a)
S = sum(calca(n) for n in range(1, 10000))
print("sum S =", S)

# (b)
a_val = 16 
Sa = sum(calca(n) for n in range(1, a_val+1))
error = abs(S - Sa)
print(f"Partial sum Sa for a={a_val}:", Sa)
print(f"Error |S - Sa|:", error)

# (c)
Sa100 = sum(calca(n) for n in range(1, a_val+101))
err100 = abs(S - Sa100)
print(f"Error for Sa+100:", err100)

Sa1000 = sum(calca(n) for n in range(1, a_val+1001))
err1000 = abs(S - Sa1000)
print(f"Error for Sa+1000:", err1000)
```
